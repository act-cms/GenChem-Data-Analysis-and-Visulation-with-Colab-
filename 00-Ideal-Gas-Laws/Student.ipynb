# Overview
This module is designed to introduce students to practice transitioning among table - graph - mathematical models for ideal gases. Students will use Google Colab, Gemini, and Python programming language to read, interpret, and analyze provided experiemental data.

Written by Dr. Linlin Jensen

> **AI Use Disclaimer:** Microsoft Copilot was used to improve clarity and organization of the activity’s written instructions.

### Learning Objectives
By the end of this recitation, you will be able to:
* Scientific Skills
  - **Data Interpretation:** Analyze how volume changes with temperature from a data table..
  - **Graphical Representation:** Construct a volume vs. temperature graph and interpret the meaning of the slope and intercepts.
  - **Mathematical Modeling:** Derive the mathematical relationship from the graph and data. Describe the relationship between volume and temperature and connect it to the Ideal Gas Law, PV = nRT.

* Cyberinfrastructure Skills
  - **Data Handling:** Import and load a CSV file into a Pandas DataFrame.
  - **Python Programming:**
    - Access and open a notebook in Google Colab
    - Identify and use the key components of the Google Colab interface (code cells, text cells, run button).
    - Read and understand a Python script with comments.
    - Run a Python script and interpret its output.
  - **Data Cleaning/Transformation:** Create a new column by converting Celsius to Kelvin.
  - **Data Visualization:** Generate a scatter plot and a linear regression line using Python libraries (e.g., Pandas, Matplotlib, NumPy/SciPy).
  - **Computational Analysis:** Perform a linear fit to find the slope and intercept.
  - **AI Literacy:** Formulate effective and specific prompts to get the most out of an AI assistant.


# Getting Started

## Step 1: Download Relevant Files and Access Google Colab

1. Download the files "Intro_recitation_gas_law_student.ipynb" and "gas_law_simulated_data.csv" from your recitation Canvas course site. These files will most likely be stored in the Download folder of your local computer.

2.   Open your web browser and go to https://colab.research.google.com/.

3.   Log in with your Google account. This is the same account you use for Gmail, Google Drive, etc. If you don't have one, you'll need to create a free account.

## Step 2: Open a Notebook

Once you're on the Colab homepage, click "File" > "Open notebook": This will open a window with several tabs:

- "Recent": Lists recently opened Colab files.

- "Google Drive": Allows you to browse your Google Drive folders to select a notebook.

- "Github": Allows you to open a notebook directly from a GitHub repository (requires the URL or repository details).

- "Upload": Allows you to upload a .ipynb file from your local computer. You can drag and drop the file here or click "Browse", locate the notebook on your machine, and click "Open".

**Use the "Upload" option to open the "Intro_recitation_gas_law_student.ipynb" file.**


## Step 3: Tour the Interface

You should now see a Colab notebook. Here are the key parts ycou'll be using:

- Code Cells: These are the gray boxes where you'll write and run your code. You can add a new one by clicking on "+ Code" at the top of the notebook.

- Text Cells: These are for adding notes, explanations, and instructions. You can add a new one by clicking on "+ Text" at the top of the notebook.

- Run Button: To run a code cell, click the "play" button (a triangle) on the left side of the cell, or press Shift + Enter on your keyboard.

## Step 4: Upload the Experimental Data to Colab Worksheet

To achieve this, following the steps below:
1. Make sure that you have downloaded the file "gas_law_simulated_data.csv" from your recitation Canvas course site. *This file contains a series of volume measurement at different temperatures for a 0.25 mole sample of ideal gas, recorded at a constant temperature*.
2. Click on the file explorer icon on the left sidebar
3. Drag and drop the file to the sidebar. A window will pop out and give you a Warning: **Ensure that your files are saved elsewhere. This runtime's files will be deleted when this runtime is terminated.** This means that the file will only be temporarily stored in the folder. Once you close the notebook, the file will be deleted and disappear.
4. Click on "Okay" and the file should appear below the folder titled "sample_data".
